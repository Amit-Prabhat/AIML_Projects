```python
df_Car_mpg= df_Car_mpg.drop('car_name', axis =1)
```

```python
df_Car_mpg.info()
```

```python
from scipy.stats import zscore

df_Car_mpg = df_Car_mpg.loc[:, 'mpg':'origin']
df_Car_mpg_z = df_Car_mpg.apply(zscore)
```

```python
cluster_range = range( 1, 10)
cluster_errors = []
for num_clusters in cluster_range:
  kmeans = KMeans( num_clusters, n_init = 10)
  kmeans.fit(df_Car_mpg_z)
  cluster_errors.append(kmeans.inertia_ )
```

```python
plt.plot(range(1,10),cluster_errors)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('cluster_errors')
plt.show()
```

The elbow plot confirms our visual analysis that there is likely 3 clusters

```python
kmeans = KMeans( n_clusters = 3, random_state = 7 )
kmeans.fit(df_Car_mpg_z)

prediction=kmeans.predict(df_Car_mpg_z)
df_Car_mpg_z["GROUP"] = prediction
```

```python
df_Car_mpg_z.boxplot(by = 'GROUP',  layout=(2,4), figsize=(15, 10))
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist
plt.figure(figsize=(18, 16))
plt.title('Agglomerative Hierarchical Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
Z = linkage(df_Car_mpg_z, 'ward')
dendrogram(Z,leaf_rotation=90.0,p=20,color_threshold=10,leaf_font_size=10,truncate_mode='level')
plt.tight_layout()
```

In hierarchical clustering it's appropriate by interpreting the dendrogram where as in K Means clustering, since one start with random choice of clusters, the results produced by running the algorithm many times may differ.

Three optimal clusters are present in the data

```python
from sklearn.linear_model import LinearRegression
```

```python
df_Car_mpg_z_Group0= df_Car_mpg_z[df_Car_mpg_z['GROUP'] == 0]
```

```python
df_Car_mpg_z_Group1= df_Car_mpg_z[df_Car_mpg_z['GROUP'] == 1]
```

```python
df_Car_mpg_z_Group2= df_Car_mpg_z[df_Car_mpg_z['GROUP'] == 2]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X0 = df_Car_mpg_z_Group0.drop(['mpg','GROUP'], axis=1)

y0 = df_Car_mpg_z_Group0['mpg']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X0, y0, test_size=0.30, random_state=1)
```

```python
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
regression_model_pred = regression_model.predict(X_test)
```

```python
r_sq = regression_model.score(X_test, y_test)
```

```python
print('Coefficient for Group0: \n', r_sq)
```

When the clustering method is used then all the data is grouped as per the similarity. Thus it becomes easy to replace the missing data and find the hyper plane which distinguish among the data.
