```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from  tensorflow.keras.layers import Flatten, Dense
from pandas import DataFrame
from sklearn.preprocessing import PowerTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tkinter as tk
from tkinter import ttk
import pandas as pd 
import re
import warnings;
warnings.simplefilter('ignore')

main_window= tk.Tk()
main_window.title('Neural Network GUI'.center(110))
main_window.geometry("600x250")



# importing data 
Name= ttk. Label(main_window,text = "Step 1: File Name")
Name.grid(row=0, column=0)

Name_var=tk.StringVar()
Name_entrybox= ttk.Entry(main_window, width = 16, textvariable= Name_var)
Name_entrybox.grid(row=0, column=1)

def Import_data():
    global DB
    Df_Name= Name_var.get()
    DB=pd.read_csv(Df_Name)
    Confirmation ="Done"
    Confirmation_box=ttk.Entry(main_window, width=16)
    Confirmation_box.grid(row=0, column=3)
    Confirmation_box.insert(1, str(Confirmation))

Import_DataButton=ttk.Button(main_window, text='Import Data', command = Import_data).grid(row=0, column=2)


# importing Target
Name1= ttk. Label(main_window,text = "Step 2: Target Column")
Name1.grid(row=1, column=0)

Name1_var=tk.StringVar()
Name1_entrybox= ttk.Entry(main_window, width = 16, textvariable= Name1_var)
Name1_entrybox.grid(row=1, column=1)

def Import_target():
    target_Name= Name1_var.get()
    c=0
    for col in DB.columns:
        if (col == target_Name):
            Confirmation1 ="Found"
            c=1
    if (c == 0):
        Confirmation1 ="Not Found"  
        
    Confirmation1_box=ttk.Entry(main_window, width=16)
    Confirmation1_box.grid(row=1, column=3)
    Confirmation1_box.insert(1, str(Confirmation1))

Import_TargetButton=ttk.Button(main_window, text='Import Target', command = Import_target).grid(row=1, column=2)


#------------------------------------Step3-----------------------------------------------
# Neural Network Regressor
Name2= ttk. Label(main_window,text = "Step 3: Neural Network Regressor")
Name2.grid(row=2, column=0)

# Neural Network Regressor Train
Name3= ttk. Label(main_window,text = "Regression")
Name3.grid(row=3, column=0)

def Train_model():
    #Prepare Data
    x=DB.drop(['Signal_Strength'],axis=1)
    y=DB['Signal_Strength']

    from sklearn.model_selection import train_test_split
    x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.25, random_state=42)


    #Removing the outliers
    from pandas import DataFrame
    from sklearn.preprocessing import PowerTransformer
    clmn=x_train.columns
    pt = PowerTransformer()
    x_train = pt.fit_transform(x_train)
    x_train = DataFrame(x_train)
    x_train.columns=clmn

    #Removing the outliers
    from pandas import DataFrame
    from sklearn.preprocessing import PowerTransformer
    clmn=x_test.columns
    pt = PowerTransformer()
    x_test = pt.fit_transform(x_test)
    x_test = DataFrame(x_test)
    x_test.columns=clmn

    # Standard Scalar
    sc=StandardScaler()

    Xnew_train= sc.fit_transform(x_train)
    Xnew_test=sc.transform(x_test)

    ynew_train = y_train.to_numpy()
    ynew_test=y_test.to_numpy()
    
    
    # Initialize Sequential model
    global model_reg
    model_reg = tf.keras.Sequential()
    # Normalize input data
    model_reg.add(tf.keras.layers.BatchNormalization(input_shape=(11,)))
    model_reg.add(tf.keras.layers.Dense(300, activation= "relu"))
    model_reg.add(tf.keras.layers.Dense(121, activation= "relu"))
    model_reg.add(tf.keras.layers.Dropout(0.25))
    # Add final Dense layer for prediction
    model_reg.add(tf.keras.layers.Dense(1))
    model_reg.compile(optimizer='adam', loss='mse',metrics=['mae'])
 
    #Fit model
    model_reg.fit(Xnew_train, ynew_train, validation_data=(Xnew_test,ynew_test),epochs=50, batch_size=10)
    
    
    Confirmation3 ="Network Trained"  
        
    Confirmation3_box=ttk.Entry(main_window, width=20)
    Confirmation3_box.grid(row=3, column=2)
    Confirmation3_box.insert(1, str(Confirmation3))

Train_Button=ttk.Button(main_window, text='Train', command = Train_model).grid(row=3, column=1)

# Neural Network Regressor Pickle
Name4= ttk. Label(main_window,text = "Pickle")
Name4.grid(row=4, column=0)

def Save_model():
    
    model_reg.save("signal_strength_quality_regressormodel.h5")
    
    Confirmation4 ="Saved model to disk"  
        
    Confirmation4_box=ttk.Entry(main_window, width=20)
    Confirmation4_box.grid(row=4, column=2)
    Confirmation4_box.insert(1, str(Confirmation4))

Save_Button=ttk.Button(main_window, text='Run', command = Save_model).grid(row=4, column=1)


#-----------------------------Preparing data for classifier------------------
def Prepare_class_data():
    Prepare_data()
    from tensorflow.keras.utils import to_categorical
    yc = to_categorical(y, num_classes= 9)
    Xc_train, Xc_test, yc_train, yc_test = train_test_split(x, yc, test_size=.30, random_state=42)


#------------------------------------Step4-----------------------------------------------
# Neural Network Classifier
Name5= ttk. Label(main_window,text = "Step 4: Neural Network Classifier")
Name5.grid(row=5, column=0)

# Neural Network Classifier Train
Name6= ttk. Label(main_window,text = "Classifier")
Name6.grid(row=6, column=0)

def Train6_model():
    x=DB.drop(['Signal_Strength'],axis=1)
    y=DB['Signal_Strength']

    from sklearn.model_selection import train_test_split
    from tensorflow.keras.utils import to_categorical
    Xc_train, Xc_test, yc_train, yc_test = train_test_split(x, y, test_size=.30, random_state=42)
    yc_train_oh = to_categorical(yc_train, num_classes= 10)
    yc_test_oh = to_categorical(yc_test, num_classes= 10)


    #Removing the outliers
    from pandas import DataFrame
    from sklearn.preprocessing import PowerTransformer
    clmn=Xc_train.columns
    pt = PowerTransformer()
    Xc_train = pt.fit_transform(Xc_train)
    Xc_train = DataFrame(Xc_train)
    Xc_train.columns=clmn

    #Removing the outliers
    from pandas import DataFrame
    from sklearn.preprocessing import PowerTransformer
    clmn=Xc_test.columns
    pt = PowerTransformer()
    Xc_test = pt.fit_transform(Xc_test)
    Xc_test = DataFrame(Xc_test)
    Xc_test.columns=clmn

    # Standard Scalar
    sc=StandardScaler()

    Xnew_train= sc.fit_transform(Xc_train)
    Xnew_test=sc.transform(Xc_test)

    
    global model_class
    model_class = tf.keras.models.Sequential()

    model_class.add(tf.keras.layers.BatchNormalization())
    model_class.add(tf.keras.layers.Dense(400, activation='relu'))
    model_class.add(tf.keras.layers.Dense(150, activation='relu'))
    model_class.add(tf.keras.layers.Dropout(0.25))
    model_class.add(tf.keras.layers.Dense(10, activation='softmax'))
    
    earlystopping= tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience= 2,restore_best_weights= True)
    
    # Compile the model
    model_class.compile(loss="categorical_crossentropy", metrics=["accuracy"], optimizer="adam")

    # Fit the model
    model_class.fit(x=Xc_train, y=yc_train_oh, validation_data=(Xc_test,yc_test_oh), epochs=50, batch_size=32, callbacks=earlystopping) 
    
    Confirmation6 ="Network Trained"  
        
    Confirmation6_box=ttk.Entry(main_window, width=20)
    Confirmation6_box.grid(row=6, column=2)
    Confirmation6_box.insert(1, str(Confirmation6))

Train6_Button=ttk.Button(main_window, text='Train', command = Train6_model).grid(row=6, column=1)

# Neural Network Classifier Pickle
Name7= ttk. Label(main_window,text = "Pickle")
Name7.grid(row=7, column=0)

def Save7_model():
    model_class.save("signal_strength_quality_Classificationmodel.h5")
    
    Confirmation7 ="Saved model to disk"  
        
    Confirmation7_box=ttk.Entry(main_window, width=20)
    Confirmation7_box.grid(row=7, column=2)
    Confirmation7_box.insert(1, str(Confirmation7))

Save7_Button=ttk.Button(main_window, text='Run', command = Save7_model).grid(row=7, column=1)

    


main_window.mainloop()
```
